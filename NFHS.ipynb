```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Import the Healthcare dataset from an Excel file
df = pd.read_excel("Healthcare.xls")
```

```python
# Assuming that you have a target variable and feature variables
target_column_name = 'Men age 15 years and above who use any kind of tobacco (%)'
feature_columns = ['Women age 15 years and above who consume alcohol (%)']
```

```python
# Separate the features (X) and the target variable (y)
X = df[feature_columns]
y = df[target_column_name]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75)
# Create a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the test data
predictions = model.predict(X_test)

# Evaluate the model
print("Mean squared error:", mean_squared_error(y_test, predictions))
print("R-squared:", r2_score(y_test, predictions))

```

```python
# Plot the data points and the regression line
plt.scatter(X_test, y_test, color='blue', label='Actual Data')
plt.plot(X_test, predictions, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Women who consume alcohol (%)')
plt.ylabel('Men who use tobacco (%)')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
